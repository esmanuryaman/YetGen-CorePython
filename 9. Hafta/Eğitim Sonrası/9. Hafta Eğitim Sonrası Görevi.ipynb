# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
import random
import time

class Kumanda:
    def __init__(self, tvDurumu='Kapalı', tvSes=0, kanalListesi=None, mevcutKanal=''):
        self.tvDurumu = tvDurumu
        self.tvSes = tvSes
        self.kanalListesi = kanalListesi if kanalListesi else []
        self.mevcutKanal = mevcutKanal

    def tvAc(self):
        if self.tvDurumu == 'Açık':
            print("Televizyon zaten açık.")
        else:
            self.tvDurumu = 'Açık'
            print("Televizyon açıldı.")

    def tvKapat(self):
        if self.tvDurumu == 'Kapalı':
            print("Televizyon zaten kapalı.")
        else:
            self.tvDurumu = 'Kapalı'
            print("Televizyon kapatıldı.")

    def sesAyarları(self):
        while True:
            cevap = input("Sesi Azalt: '-' \nSesi Artır: '+' \nÇıkış: 'q'\nSeçiminiz: ")

            if cevap == '-':
                if self.tvSes > 0:
                    self.tvSes -= 1
                    print(f"Ses: {self.tvSes}")
            elif cevap == '+':
                if self.tvSes < 31:
                    self.tvSes += 1
                    print(f"Ses: {self.tvSes}")
            elif cevap == 'q':
                print("Ses ayarlarından çıkılıyor.")
                break
            else:
                print("Geçersiz işlem.")

    def kanalEkle(self, kanalIsmi):
        print("Kanal ekleniyor...")
        time.sleep(1)
        self.kanalListesi.append(kanalIsmi)
        print("Kanal eklendi.")

    def kanalSayısınıOgrenme(self):
        print(f"Kanal sayısı: {len(self.kanalListesi)}")

    def rastgeleKanalaGeçme(self):
        if self.kanalListesi:
            self.mevcutKanal = random.choice(self.kanalListesi)
            print(f"Şu anki kanal: {self.mevcutKanal}")
        else:
            print("Kanal listesi boş.")

    def televizyonBilgileri(self):
        print(f"Televizyon Durumu: {self.tvDurumu}")
        print(f"Ses: {self.tvSes}")
        print(f"Kanal Listesi: {self.kanalListesi}")
        print(f"Mevcut Kanal: {self.mevcutKanal}")

def main():
    kumanda = Kumanda()

    print("""
    Televizyon Uygulaması

    1. Tv Aç
    2. Tv Kapat
    3. Ses Ayarları
    4. Kanal Ekle
    5. Kanal Sayısını Öğrenme
    6. Rastgele Kanala Geçme
    7. Televizyon Bilgileri

    Çıkmak için 'q' ya basın.
    """)

    islemler = {
        '1': kumanda.tvAc,
        '2': kumanda.tvKapat,
        '3': kumanda.sesAyarları,
        '4': lambda: [kumanda.kanalEkle(kanal.strip()) for kanal in input("Kanal isimlerini ',' ile ayırarak giriniz: ").split(',')],
        '5': kumanda.kanalSayısınıOgrenme,
        '6': kumanda.rastgeleKanalaGeçme,
        '7': kumanda.televizyonBilgileri
    }

    while True:
        islem = input("İşlemi Seçiniz: ")

        if islem == 'q':
            print("Programdan çıkılıyor...")
            break
        elif islem in islemler:
            islemler[islem]()
        else:
            print("Geçersiz işlem.")

if __name__ == "__main__":
    main()
```
